```python
import pandas as pd
emails = pd.read_csv("TALN/emails.csv")
emails["text"]
```

```python
from IPython.utils.text import string 
from nltk.tokenize import word_tokenize, sent_tokenize
stop_words = stopwords.words("english")
my_list = []
for i in emails["text"]:
  words = word_tokenize(i)
  ns = []
  for j in words:
    if j not in stop_words and j not in string.punctuation:
      ns.append(j)
  my_list.append(ns)
#print(my_list)
```

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
lem_list = []
for m in my_list:
  for word in m:
    lem_list.append(lemmatizer.lemmatize(word, pos='v'))
#print(lem_list)
```

```python
import json
with open('after lemmatisation.json', 'w') as file:
    json.dump(lem_list, file)
```

```python
from nltk.stem import PorterStemmer
ps=PorterStemmer()
stem_list= []
for x in my_list:
  for y in x:
    stem_list.append(ps.stem(y))
#print(stem_list)
```

```python
import json
with open('after stemming.json', 'w') as file:
    json.dump(lem_list, file)
```

```python
import math 
from collections import Counter
tf = {}
tfs = []
tf_doc = {}
i = 0
for doc in my_list:
  i=i+1
  k = []
  counted = Counter(doc)
  for element, frequency in counted.items():
      #print(f"{element}: {frequency}")
      tf.update({element:frequency})
      k.append(frequency)
  tfs.append(k)
  tf_doc.update({'D'+str(i):tf})
print(tf_doc)

```

```python
import json
with open('TF.json', 'w') as file:
    json.dump(tf_doc, file)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
#print(emails["text"])
tab = []
for i in emails["text"]:
  tab.append(i)

#
# create a TfidfVectorizer object
vectorizer = TfidfVectorizer()

# fit the vectorizer to the documents and transform the documents into TF-IDF vectors
tfidf_vectors = vectorizer.fit_transform(tab)
#print("Vocabulary:", vectorizer.get_feature_names_out())

result = pd.DataFrame(tfidf_vectors.T.todense(), index=vectorizer.get_feature_names_out(), columns=["doc"+str(i+1) for i in range(len(tab))])

#result = pd.DataFrame(tfidf_vectors.toarray(),)
print(result)
```

```python
import numpy as np

# create a 3x3 matrix
matrix = np.array(result)

# save the matrix to a file
np.savetxt('matrix.txt', matrix)
```
